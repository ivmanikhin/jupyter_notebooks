```python
import json
import re

with open("March.json", "r", encoding='utf-8') as file:
    messages = json.load(file)
messages = messages["messages"]

# print(messages)

pretty_messages = list()
for message in messages:
    try:
        pretty_messages.append({"user": message['from'], "text": message["text"], "price": re.findall(r" *([0-9.]+) —Ä.", message["text"])[0]})
    except:
        pass

pretty_messages
```
